### **Autenticação**

Para que a API criada tenha acesso a API WEB Spotify é necessário realizar uma autenticação. Para isso, ainda no Dashboard de sua API acesse setting> Basic Information e obtenha as credenciais para se comunicar com a Web API copiando e colando copie o valor do **CLIENT ID** e **CLIENT SECRET** para serem utilizados na autenticação.

image.png


A autenticação pode ser feita através de uma requisição, conforme o exemplo abaixo em **curl** :
```
CLIENT_ID="SEU_CLIENT_ID"
CLIENT_SECRET="SEU_CLIENT_SECRET"

# Codifica o cliente_id e o cliente_secret como base64
AUTH=$(printf "%s:%s" "$CLIENT_ID" "$CLIENT_SECRET" | base64)

# Faz a requisição para obter o token de acesso
TOKEN=$(curl -X "POST" -H "Authorization: Basic $AUTH" -d "grant_type=client_credentials" "https://accounts.spotify.com/api/token")

echo $TOKEN
```



Exemplo de autenticação utilizando python :

````
import base64
import requests

def autenticacao(client_id, client_secret):
  auth_str = f"{client_id}:{client_secret}"
  base64_auth_str = base64.b64encode(auth_str.encode()).decode('utf-8')

  headers = {
      'Authorization': f'Basic {base64_auth_str}'
  }


  payload = {
      'grant_type': 'client_credentials'
  }

  token_url = 'https://accounts.spotify.com/api/token'
  response = requests.post(token_url, headers=headers, data=payload)

  if response.status_code == 200:
      token_data = response.json()
      access_token = token_data['access_token']
      return access_token
  else:
      print('Error obtaining access token')
````

````
client_id = 'SEU_CLIENT_ID'
client_secret = 'SEU_CLIENT_SECRET'
access_token = autenticacao(client_id, client_secret)
print(f'Access Token: {access_token}')
````

Além de informações gerais, é possível consultar o catálogo de álbuns publicado por um artista dentro da plataforma

###**Buscando novos lançamentos**

É possível consultar uma lista com os últimos conteúdos publicados na plataforma, que pode ser muito útil aos usuários que gostam de estar sempre atualizados.

Detalhes

O endpoint Get New Releases disponível na URL https://api.spotify.com/v1/browse/new-releases possibilita a consulta a uma lista com as músicas publicadas mais recentemente, através das seguintes informações:
*  Token de autenticação: informado no corpo da requisição
*  País: informado na URL, é uma informação opcional que pode ser usada para destinar uma busca com informações mais relevantes para uma determina região
*  Limite: informa a quantidade de itens retornado na pesquisa (limitado a 50). É uma informação opcional e, quando não informada, o limite '20' será adotado pela aplicação
*  offset: este campo possibilita iniciar a contagem a partir do elemento x na lista das publicações mais recentes. É uma informação opcional e, quando não informado, o valor '0' será adotado pela aplicação





Exemplo utilizando curl:
```
curl --request GET \
  --url https://api.spotify.com/v1/browse/new-releases \
  --header 'Authorization: Bearer 1POdFZRZbvb...qqillRxMr2z'
```



Exemplo em python

````
import requests

# Constroi a URL para o endpoint de novos lancamentos
url = 'https://api.spotify.com/v1/browse/new-releases'
headers = {'Authorization': f'Bearer {access_token}'}
params = {'country': 'US', 'limit': 5, 'offset': 0}  # Você pode ajustar esses parâmetros conforme necessário

# Faz a solicitacao GET para obter os novos lancamentos
response = requests.get(url, headers=headers, params=params)

if response.status_code == 200:
    new_releases_data = response.json()

    # Exibe informacoes sobre os novos lancamentos
    for release in new_releases_data['albums']['items']:
        print('Nome do Álbum:', release['name'])
        print('Artistas:', ', '.join([artist['name'] for artist in release['artists']]))
        print('Link para o Álbum:', release['external_urls']['spotify'])
        print('\n')
else:
    print('Algo deu errado ao obter novos lançamentos')
    print(response.json())
````
